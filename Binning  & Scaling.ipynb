# Binning 

##### Binning data will convert data into discrete buckets, allowing you to gain insight into your data in logical ways

##### cut() and qcut() from pandas

cut() => Bin values into discrete intervals.

Use cut when you need to segment and sort data values into bins. This function is also useful for going from a continuous 

variable to a categorical variable.

Internally it contains a categories array specifying the disâ€ tinct category names along with a labeling for the ages data in the codes attribute

qcut() => bins the data based on sample quantiles. Dependingon the distribution of the data, using cut will not usually result 
in each bin having thesame number of data points. Since qcut uses sample quantiles instead, by definitionyou will obtain roughly equal-size bins

##### Pandas cut: Binning Data into Custom Bins

```python
# Loading a Sample Pandas DataFrame
import pandas as pd

df = pd.DataFrame.from_dict({
    'Name': ['Ray', 'Jane', 'Kate', 'Nik', 'Autumn', 'Kasi', 'Mandeep', 'Evan', 'Kyra', 'Jim'],
    'Age': [12, 7, 33, 34, 45, 65, 77, 11, 32, 55]
})

print(df.head())
```

```python
df['Age Group'] = pd.cut(
   df['Age'], 
   [0, 17, 64, 100]
)
print(df.head())
```

```python
# Adding labels to the groupings
df['Age Group'] = pd.cut(
    df['Age'], 
    [0, 17, 64, 100], 
    labels=['0-18 years old', '18-65 years old', '65+ years old']
)
print(df.head())
```

```python
# Using the right= argument to modify binning behavior
df['Age Group'] = pd.cut(
    df['Age'], 
    [0, 18, 65, 100], 
    labels=['0-18 years old', '18-65 years old', '65+ years old'],
    right=False
)
print(df.head())
```

##### Pandas qcut: Binning Data into Equal-Sized Bins

```python
# Splitting Age Column into Four Quantiles
df['Age Groups'] = pd.qcut(df['Age'], 4)
print(df.head())
```

```python
# Checking the data type of the qcut column
df['Age Groups'] = pd.qcut(df['Age'], 4)
print(df['Age Groups'].dtype)
```

##### Splitting Data Into Equal Percentiles Using Pandas qcu

```python
#split our data into 25% bins.
# Splitting Age Column into Four Quantiles
df['Age Groups'] = pd.qcut(
   df['Age'], 
   [0, 0.25, 0.5, 0.75, 1]
)
print(df.head())
```

```python
# Adding Labels to Pandas .qcut()
df['Age Groups'] = pd.qcut(
   df['Age'], 
   [0, 0.25, 0.5, 0.75, 1], 
   labels=['0-25%', '26-49%', '51-75%', '76-100%']
)
print(df.head())
```

```python
# Modifying Precision in Categories
# by default: precision=3
df['Age Groups'] = pd.qcut(
   df['Age'], 
   4, 
   precision=0
)
print(df.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
x = np.arange(0,20,0.4)
y1 = np.sin(x)
y2 = np.cos(x)*50
```

```python
plt.plot(x,y1,'red')
plt.plot(x,y2,'blue')
```

```python
#Scaling
y1_new = y1/max(y1)
y2_new = y2/max(y2)
plt.plot(x,y1_new,'red')
plt.plot(x,y2_new,'blue')
```

```python
#### Min-Max Scaling
```

```python
y1_new = (y1-min(y1))/(max(y1)-min(y1))
y2_new = (y2-min(y2))/(max(y2)-min(y2))
plt.plot(x,y1_new,'red')
plt.plot(x,y2_new,'blue')
```

```python
##### Normalization
```

```python
y1_new = (y1-np.mean(y1))/(max(y1)-min(y1))
y2_new = (y2-np.mean(y2))/(max(y2)-min(y2))
plt.plot(x,y1_new,'red')
plt.plot(x,y2_new,'blue')
```

```python
##### Standaridization
```

```python
y1_new = (y1-np.mean(y1))/np.std(y1)
y2_new = (y2-np.mean(y2))/np.std(y2)
plt.plot(x,y1_new,'red')
plt.plot(x,y2_new,'blue')
```
